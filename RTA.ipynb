```python
plt.figure(figsize=(20, 12))
accidents_plot = sns.barplot(x='location', y='Total number of accidents', hue='Year', data=pd, palette='pastel')
accidents_plot.set_xticklabels(accidents_plot.get_xticklabels(), rotation=0, ha="center")
```
