```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.io as pio
import plotly.express as px
```

```python
pio.renderers.default = 'iframe_connected'
```

```python
data=pd.read_csv('traffic-accidents-annual-.csv')
data
```

```python
data.info()
```

```python
data.head()
```

```python
data.columns
```

```python
data.rename(columns={'*- Islamabad Included.':'Place','Unnamed: 1':'Year','Unnamed: 2':'Total_accidents','Unnamed: 3':'Fatal_Accident',
                    'Unnamed: 4':'Non-Fatal_Accidents','Unnamed: 5':'Killed','Unnamed: 6':'Injured','Unnamed: 7':'Total_vehicles_involved'},inplace=True)
```

```python
data
```

- Drop First Row

```python
data=data.iloc[1:]
data
```

```python
data.head(10)
```

- Some Years written with (*) so it should be replace

```python
data['Year']=data['Year'].replace("2012-13 * ","2012-13")
data['Year']=data['Year'].replace("2013-14*","2013-14")
data['Year']=data['Year'].replace("2014-15*","2014-15")
data['Year']=data['Year'].replace("2015-16*","2015-16")
data['Year']=data['Year'].replace("2016-17*","2016-17")
```

```python
data.head(10)
```

- (-)sign includes in year columns it should be split

```python
data['Year']=data['Year'].str.split("-").str[0]
data=data.explode('Year',ignore_index=True)
```

```python
data.head()
```

- Change data types of Columns

```python
data=data.astype({'Place':str,'Year':int,'Total_accidents':int,'Fatal_Accident':int,'Non-Fatal_Accidents':int,
                 'Killed':int,'Injured':int,'Total_vehicles_involved':int})
data
```

```python
fa=data[{'Year','Fatal_Accident','Place'}].loc[(data.Place=='Pakistan')]
graph=px.bar(fa,x='Year',y='Fatal_Accident',color='Year',text='Fatal_Accident',title='Total Fatal Accident in Pakistan')
graph.update_traces(texttemplate='%{text}',textposition='outside')
graph.show()
```

```python
pa=data[{'Year','Total_accidents'}].loc[(data.Place=='Punjab')]
graph=px.scatter(pa,x='Year',y='Total_accidents',color='Year',size='Total_accidents',title='Total Accidents In Punjab')
graph.show()
```

```python
Place=['Sindh','Punjab',"Balochistan","Khyber Pakhtunkhwa"]
ratio=data[{'Place','Total_accidents'}].loc[data['Place'].isin(Place)]

fig=px.pie(ratio,values='Total_accidents',names='Place',width=900,height=900,color='Place',color_discrete_map={'Punjab':'red',
                                 'Sindh':'orange',
                                 'Balochistan':'royalblue',
                                 'Khyber Pakhtunkhwa':'green'},
                                 title='Ratio Of Traffic Accident in Provinces ')
fig.show()
```

```python
Place=['Sindh','Punjab',"Balochistan","Khyber Pakhtunkhwa"]
dr=data[{'Place','Killed'}].loc[data['Place'].isin(Place)]

fig=px.pie(dr,values='Killed',names='Place',width=950,height=900,color='Place',color_discrete_map={'Punjab':'lightcyan',
                                 'Sindh':'cyan',
                                 'Balochistan':'royalblue',
                                 'Khyber Pakhtunkhwa':'darkblue'},
                                 title='Ratio Of People Died Through Accident in Provinces ')
fig.show()
```
